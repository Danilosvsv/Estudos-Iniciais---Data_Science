# Matplotlib

- Biblioteca de visualização de dados.
- Essa biblioteca permite ao programador, controlar os aspectos da figura.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = np.exp(x)
y
```

```python
# Funcional

plt.plot(x,y,'b--')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico X x Y')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
figura = plt.figure()

ax1 = figura.add_axes([0.2,0.2,0.9,0.9])
ax2 = figura.add_axes([0.35,0.6,0.4,0.4])
ax1.plot(x,y)
ax1.set_xlabel('Eixo x')
ax1.set_ylabel('Eixo y')
ax2.set_xlabel('Eixo x')
ax2.set_ylabel('Eixo y')
ax1.set_title('Estudo com a exponencial no matplotlib')
ax2.plot(y,x)
```

Na figura abaixo é utilizada algumas das propriedades da função .plot(). Do código, temos inicialmente que no comando da linha 4 não é utilizado um marcador, porém o estilo de linha é tracejado e a cor vermelha. Já no comando da linha 5 é utilizado o marcador de ponto, o estilo de linha sólido e a cor azul. Por fim, na linha 6 é utilizado a cor amarela, o marcador de círculos e não há estilo de linha. 

```python
## Também é possível plotar o gráfico de várias funções em um único gráfico:

fig, ax = plt.subplots(figsize=(10,9))
ax.plot(x,np.sqrt(x),'r--',label='$\sqrt{x}$')
ax.plot(x,x**2,'b-.',label='$x^2$')
ax.plot(x,-x,'yo',label='$-x$',color='#54a176') # definindo cor através do HEX
ax.set_title('Gráficos de algumas funções')
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')
ax.legend(loc=0)
```

Ressalto, que essas são apenas alguns exemplos de estilos de marcação, linha e cores que podem ser utilizados para a construção dos gráficos, pois há uma gama de possibilidades que podem ser encontradas na documentação da função no próprio site da matplolib. Além disso, o usuário pode personalizar novos estilos conforme preferir.

```python
# Definindo fonte

from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['DejaVu Sans']
```

```python
fig, ax = plt.subplots(nrows = 2, ncols=2,figsize=(12,8))
ax[0,0].plot(x,np.sqrt(x),'r--')
ax[0,0].set_title('Gráfico de $y=\sqrt{x}$')
ax[0,1].plot(x,x,'b-.')
ax[0,1].set_title('Gráfico da reta $y=x$')
ax[1,0].plot(x,x**2,'b-.')
ax[1,0].set_title('Gráfico da parábola  $y=x^2$')
ax[1,1].plot(x,x**3,'b-.')
ax[1,1].set_title('Gráfico de $y=x^3$')
plt.show()
# outra forma de operacionar a construção desses gráficos é utilizando a estrutura for.
```

```python
# O comando abaixo salva a figura no mesmo diretório desse arquivo na sua máquina

fig.savefig("Graficos.png") 
```

##### Exemplo com Gráfico de Barras

```python
atletas = ('Danilo', 'Daniel', 'Maria', 'Flávia', 'Joe','Marie')
print(atletas)
pontos =  [5.787647, 5.65646916, 4.29922856, 0.50641433, 8.25207979,7.14826714]
pontos
```

```python
plt.bar(atletas, pontos, color="blue")
plt.xticks(atletas)
plt.ylabel('Pontos')
plt.xlabel('Atletas')
plt.title('Desempenho de Atletas na competição')
```

O gráfico também pode ser visualizado de forma horizontal da seguinte forma:

```python
plt.barh(atletas, pontos, color="black")
plt.ylabel('Pontos')
plt.xlabel('Atletas')
plt.title('Desempenho de Atletas na competição')
```

E se o conjunto de dados fosse diferente e a plotagem necessita-se considerar uma margem de erro?

```python
margem_erro =  [0.787647, 0.65646916, 0.29922856, 0.050641433, 0.25207979,0.14826714]
plt.barh(atletas, pontos, color="pink",xerr=margem_erro)
plt.ylabel('Pontos')
plt.xlabel('Atletas')
plt.title('Desempenho de Atletas na competição')
```

##### Exemplo com Histograma

```python
np.random.seed(42)
dados = np.random.randn(1000)
plt.hist(dados, density=True, bins=25)  # `density=False` would make counts
plt.ylabel('Probabilidade')
plt.xlabel('Dados')
plt.title("Histograma")
```
